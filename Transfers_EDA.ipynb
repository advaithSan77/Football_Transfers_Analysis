```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("football_players.csv")
```

```python
df.head()
```

```python
df
```

```python
df.isna()
```

```python
df.isna().sum()
```

#Players from each country

```python
coun = pd.DataFrame(df.groupby('Origin', as_index=False, sort=True)['Player'].count())
coun.sort_values(by=['Player'], ascending=False, inplace=True)
coun
```

```python
graph = plt.figure(figsize=(10, 5))
plt.bar(coun['Origin'], coun['Player'], color='blue', width=0.5)
plt.xlabel('Origin')
plt.ylabel('No of players')
plt.title('No of players from each country')
plt.xticks(rotation=75)
plt.show
```

#Fee for each player

```python
player_fee = df[['Player', 'Fee(€ mln)']]
player_fee.sort_values(by=['Fee(€ mln)'], ascending=False, inplace=True)
player_fee
```

```python
graph1 = plt.figure(figsize=(20, 7))
plt.bar(player_fee['Player'], player_fee['Fee(€ mln)'], color='red', width=0.4)
plt.xlabel('Player')
plt.ylabel('Price (in €)')
plt.title('Price of each player')
plt.xticks(rotation=80)
plt.show
```

#Club revenue

```python
rev = pd.DataFrame(df.groupby('From(Club)', as_index=False, sort=True)['Fee(€ mln)'].sum())
rev.sort_values(by = ['Fee(€ mln)'], ascending=False, inplace=True)
rev
```

```python
graph2 = plt.figure(figsize=(10, 5))
plt.bar(rev['From(Club)'], rev['Fee(€ mln)'], color='magenta', width=0.4)
plt.xlabel('Clubs')
plt.ylabel('Revenue(in €)')
plt.title('Revenue of each club')
plt.xticks(rotation=80)
plt.show
```

#Club expenditure

```python
exp = pd.DataFrame(df.groupby('To(Club)', as_index=False, sort=True)['Fee(€ mln)'].sum())
exp.sort_values(by = ['Fee(€ mln)'], ascending=False, inplace=True)
exp
```

```python
graph3 = plt.figure(figsize=(10, 5))
plt.bar(exp['To(Club)'], exp['Fee(€ mln)'], color='orange', width=0.4)
plt.xlabel('Clubs')
plt.ylabel('Expenditure(in €)')
plt.title('Expenditure of each club')
plt.xticks(rotation=80)
plt.show
```

#Revenue for each position

```python
df['Position'].unique()
```

```python
df['Position'] = df['Position'].apply(lambda x : x if x != 'Striker' else 'Forward')
```

```python
pos = pd.DataFrame(df.groupby('Position', as_index=False, sort=True)['Fee(€ mln)'].sum())
pos.sort_values(by=['Fee(€ mln)'], ascending=False, inplace=True)
pos
```

```python
graph4 = plt.figure(figsize=(10, 5))
plt.barh(pos['Position'], pos['Fee(€ mln)'], color='black')
plt.xlabel('Revenue (in €)')
plt.ylabel('Position')
plt.title('Total revenue for each position')
plt.show
```

#Trend of fee spent

```python
year_fee = pd.DataFrame(df.groupby('Year', as_index=False, sort=True)['Fee(€ mln)'].sum())
year_fee
```

```python
graph4 = plt.figure(figsize=(10, 5))
plt.plot(year_fee['Year'], year_fee['Fee(€ mln)'], color='black')
plt.xlabel('Year')
plt.ylabel('Money spent(€ mln)')
plt.title('Total revenue for each position')
plt.show
```

#Preferred Club

```python
pref_club = pd.DataFrame(df.groupby('To(Club)', as_index=False, sort=True)['Player'].count())
pref_club.sort_values(by=['Player'], ascending=False, inplace=True)
pref_club
```

```python
graph4 = plt.figure(figsize=(10, 5))
plt.bar(pref_club['To(Club)'], pref_club['Player'], color='black', width=0.4)
plt.xlabel('To(Club)')
plt.ylabel('Number of players')
plt.title('Preferred clubs by players')
plt.xticks(rotation=70)
plt.show
```
